```python
import os
path =  "vanilla" #folder with files to rewrite

rootList = os.listdir(path)
rootList
```

```python
class Node: 
    def __init__(self, name, dire, child): 
        self.name = name
        self.path = dire
        self.child = {}
        
        for name in child: #will create a tree with all files and folders from the root node directory
            self.child[name] = []
            try:
                path = dire+"/"+name
                listInner = os.listdir(path)
                
                leaf = Node(name,path,listInner)
                self.child[name].append(leaf)
            except:
                pass
    
    def __str__(self):
        return self.name + "//" + str(len(self.child))
    
    def __repr__(self):
        return self.name + "//" + str(len(self.child))
```

```python
import re
def extract_numbers(string): #extracts both float and integers, ignoring country tags
    floats = re.findall("\d+\.\d+", string)
    string = filter_list(string,floats)
    tags = re.findall("[a-zA-Z]\d\d", string)
    string = filter_list(string,tags)
    ints = re.findall("\d+", string)
    
    return floats+ints
```

```python
def single_line_multimod(line): #making sure to deal with all possible edge cases
    line = line.replace("="," = ") 
    line = line.replace("{"," { ")
    line = line.replace("}"," } ")
    
    blocks = list(filter(len,line.split(" ")))
    blocks.append("")
    blocks.append("")
    newline = ""
    j = 0
    for i in range(len(blocks) - 2):
        j = j - 1
        for x in multiplier:
            if(x in blocks[i] and blocks[i+1] == "=" and 
               blocks[i+2].replace('-','',1).replace('.','',1).isdecimal()):
                
                newline = newline + blocks[i] + " = " + str(float(blocks[i+2])*multiplier[x]) + " "
                j = 2
                break
        if(j < 0 ):
            newline = newline + blocks[i] + " "
    return newline
```

```python
import shutil
def copy_file(filePath, fileName, writePath):
    if not os.path.exists(writePath):
            os.makedirs(writePath)
    shutil.copyfile(filePath+ "/" + fileName, writePath + "/" + fileName)
```

```python
def rewrite_file(filePath, fileName, writePath):
    with open(filePath + "/" + fileName, "rt", encoding = "cp1252", errors = "replace") as fin:
        if not os.path.exists(writePath):
            os.makedirs(writePath)
        with open(writePath + "/" + fileName, "wt", errors = "replace") as fout:
            try:
                for line in fin:
                    line, tag, comms = line.partition("#") #escape comments
                    comms = tag+comms

                    numbers = extract_numbers(line)

                    #Don't multiply estate task and privileges requirements
                    if("on_granted = {" in line or "can_select = {" in line or 
                         "pre_effect = {" in line or "task_requirements = {" in line):
                        i = line.count("{") - line.count("}")
                        fout.write(line+comms)
                        while i:
                            line = fin.readline()
                            line, tag, comms = line.partition("#")
                            comms = tag+comms
                            i = i + line.count("{") - line.count("}")
                            fout.write(line+comms)
                        
                    #Don't change lines with no modifiers/mismatched substrings    
                    elif(len(numbers) == 0 or "Arangerel" in line or "legitimacy_equivalent" in line):
                        fout.write(line+comms)

                    else:
                        match = False
                        for x in multiplier: #check which modifier/multiplier
                                if(x in line):
                                    match = True
                                    if(len(numbers) == 1): #multiplying single modifier
                                        split = line.split(numbers[0])
                                        number = float(numbers[0])*multiplier[x]
                                        newline = split[0]+str(number)+split[-1]

                                        fout.write(newline+comms)

                                    if(len(numbers) > 1): #treating multiple modifiers in a single line 
                                        newline = single_line_multimod(line)
                                        fout.write(newline+comms)
                                    break
                        if(not match): #No modifier matched, write line with no changes
                            fout.write(line+comms)
            except:
                print(fileName,line)
                    
def rewrite_folder(node):
    for (x,y) in node.child.items():
        
        if(y):
            rewrite_folder(y[0])
        else:
            if("technologies" in node.path or "static_modifiers" in node.path or
               "crown_land" in node.path or ".lua" in x):
                copy_file(node.path,x,"testfolder/"+node.path)
            elif(".txt" in x):
                rewrite_file(node.path,x,"testfolder/"+node.path)
```
